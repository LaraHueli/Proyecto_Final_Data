```python
# importaciones
import pandas as pd
import numpy as np

import os
import sys

ruta = r"c:\Users\antonio\Desktop\Proyecto_Final_Data\src"  # Ruta a src
sys.path.append(ruta)  
import sp_limpieza
import sp_visualizaciones

from sp_limpieza import limpiar_nombres_columnas 
from sp_limpieza import limpiar_fechas
from sp_limpieza import separar_fecha_hora
```

```python
# carga y lectura de archivos(sales)
ruta = "../data/data_raw/sales.csv"
df_sales_raw = pd.read_csv(ruta)
```

```python
#creamos copia entes empezar la limpieza y transformacion
df_sales = df_sales_raw.copy()  
```

```python
print(df_sales.shape)  # Devuelve (filas, columnas)
df_sales.info() # Muestra informacion general
df_sales.head() # Muestra las primeras filas
```

```python
print(df_sales.isnull().sum())
porcentaje_nulos = (df_sales["SalesDate"].isnull().sum() / len(df_sales)) * 100
print(f"Porcentaje de valores nulos en SalesDate: {porcentaje_nulos:.2f}%")

```

```python
df_sales = df_sales.dropna(subset=["SalesDate"])
print(df_sales.shape)  # Para confirmar que los valores nulos se eliminaron
print(df_sales.isnull().sum())
```

```python
df_sales = limpiar_nombres_columnas(df_sales)
df_sales.head()
```

```python
df_sales = separar_fecha_hora(df_sales, "Sales_Date")
df_sales.head()
```

```python
df_sales.drop(columns=["Sales_Date"], inplace=True)
df_sales.head()
```

```python
print(df_sales["Transaction_Number"].nunique())  # Cantidad de valores únicos en Transaction_Number
print(df_sales["Sales_Id"].nunique())  # Cantidad de valores únicos en Sales_Id
print(df_sales.duplicated(subset=["Sales_Id", "Transaction_Number"]).sum())
```

```python
df_sales.drop(columns=["Transaction_Number"], inplace=True)
df_sales.head()
```
