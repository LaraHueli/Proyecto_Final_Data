```python
# importaciones
import pandas as pd
import numpy as np

import os
import sys

ruta = r"c:\Users\antonio\Desktop\Proyecto_Final_Data\src"  # Ruta a src
sys.path.append(ruta)  
import sp_limpieza

from sp_limpieza import limpiar_nombres_columnas 
from sp_limpieza import concatenar_nombres
from sp_limpieza import limpiar_fechas
from sp_limpieza import calcular_edad
from sp_limpieza import años_trabajados
```

```python
# carga y lectura de archivos(emplopyees)
ruta = "../data/data_raw/employees.csv"
df_employees_raw = pd.read_csv(ruta)
```

```python
#creamos copia entes empezar la limpieza y transformacion
df_employees = df_employees_raw.copy()  
```

```python
print(df_employees.shape)  # Devuelve (filas, columnas)
df_employees.info() # Muestra informacion general
df_employees.head() # Muestra las primeras filas
```

```python
# Eliminamos columna MiddleInitial porque no aporta valor
df_employees = df_employees.drop(columns=["MiddleInitial"])
df_employees.head() # Verificamos que ya no esté
```

```python
df_employees = limpiar_nombres_columnas(df_employees)
df_employees.head()
```

```python
df_employees = concatenar_nombres(df_employees)
df_employees.head()
```

```python
columnas_con_fechas = ["Birth_Date", "Hire_Date"]  # Lista de columnas con fechas
df_employees = limpiar_fechas(df_employees, columnas_con_fechas)
df_employees.head()
```

```python
df_employees = calcular_edad(df_employees, "Birth_Date")
df_employees.head()
```

```python
df_employees = años_trabajados(df_employees, "Hire_Date")
df_employees.head()
```
